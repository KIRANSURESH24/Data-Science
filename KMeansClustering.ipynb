```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
data=load_iris()
x=data.data
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(x)
pred=km.predict(x)
plt.scatter(x[pred==0,0],x[pred==0,1],c='blue',label='cluster1')
plt.scatter(x[pred==1,0],x[pred==1,1],c='red',label='cluster2')
plt.scatter(x[pred==2,0],x[pred==2,1],c='green',label='cluster3')
plt.scatter(x[pred==3,0],x[pred==3,1],c='black',label='cluster4')
plt.legend()
plt.show()
```
